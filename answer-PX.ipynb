```python
## Your code here
import numpy as np
import matplotlib.pyplot as plt
from functions import Compose, Sin, Exponential, newton_root, newton_extremum
f = Compose(Sin(),Exponential())
x0 = 1.0
x = newton_root(f,x0)
x1 = np.linspace(-2,2,100)
plt.plot(x1,f.evaluate(x1))
plt.scatter(x,f.evaluate(x))
plt.text(x, f.evaluate(x)+0.1,"root")
```

## Part B

```python
## Your code here
x0 = 0.0
x = newton_extremum(f, x0)
x1 = np.linspace(-2,2,100)
plt.plot(x1,f.evaluate(x1))
plt.scatter(x,f.evaluate(x))
plt.text(x-0.5, f.evaluate(x)+0.1,"local extremum")
```
